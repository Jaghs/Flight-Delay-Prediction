## Cleaning

### flights_sample

```python
flights_dropCols = ['no_name', 'first_dep_time', 'total_add_gtime', 'longest_add_gtime',
                'flights', 'dest_airport_id', 'origin_airport_id', 'mkt_carrier', 'tail_num',
               ]
flights_fillCols = ('carrier_delay', 'weather_delay', 'nas_delay', 'security_delay',
                    'late_aircraft_delay')
# FIXME: Slicing warnings need to be dealt with
flights_lite = flights_sample_df.drop(columns = flights_dropCols)
# BUG: the next line removes all cancellation values as a side-effect, so cancellations need
# to be engineered into a new feature that is present when there are values for arr_delay.
flights_clean = flights_lite.dropna(subset = ['arr_delay'])
flights_clean.loc[:, flights_fillCols] = flights_clean.loc[:, flights_fillCols].fillna(0)
flights_clean['cancellation_code'].fillna("0", inplace = True)
flights_clean.dropna(inplace = True)
```

```python
passengers_df.nunique()
```

### fuel_consumption

```python
fuel_dropCols = ['airline_id', 'carrier', 'carrier_name']
fuel_lite = fuel_consumption_df.drop(columns = fuel_dropCols)
fuel_clean = (fuel_lite
              .dropna(subset = ['unique_carrier'])
              .fillna(fuel_lite.mean())
             )
```

## Merging DFs for modelling

```python
# some pre-merge housekeeping
flights_clean["route_id"] = flights_clean[["origin", "dest"]].apply("-".join, axis=1)
passengers_df["route_id"] = passengers_df[["origin", "dest"]].apply("-".join, axis=1)
flights_clean['month'] = flights_clean['fl_date'].dt.month
flights_clean['year'] = flights_clean['fl_date'].dt.year
passengers_df.rename(columns = {'unique_carrier': 'mkt_unique_carrier'}, inplace = True)
fuel_clean.rename(columns = {'unique_carrier': 'mkt_unique_carrier'}, inplace = True)
```

```python
# FIXME Merge should only result in a table with no more rows than flights_clean
flights_passengers = pd.merge(flights_clean, passengers_df, on=['year', 'month', 'mkt_unique_carrier', 'route_id'], how = 'left')
```

```python
flights_passengers.shape
```

```python
flights_fuel = pd.merge(flights_clean, fuel_clean, on=['year', 'month', 'mkt_unique_carrier'], how = 'left')
```

```python
all_features = flights_fuel
all_features.to_csv('data/all_features.csv', index = False)
```
